```python
# Import the NetworkX package
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

## Graph
NetworkX provides several classes to store different types of graphs, such as directed and undirected graph. It also provides classes to create multigraphs (both directed and undirected).

For more information, please refer to [NetworkX graph types](https://networkx.org/documentation/stable/reference/classes/index.html).

```python
# Create an undirected graph G
G = nx.Graph()
print(G.is_directed())

# Create a directed graph H
H = nx.DiGraph()
print(H.is_directed())

# Add graph level attribute
G.graph["Name"] = "Bar"
print(G.graph)
```

## Node

Nodes (with attributes) can be easily added to NetworkX graphs.

```python
# Add one node with node level attributes
G.add_node(0, feature=5, label=0)

# Get attributes of the node 0
node_0_attr = G.nodes[0]
print("Node 0 has the attributes {}".format(node_0_attr))
```

```python
G.nodes(data=True)
```

```python
# Add multiple nodes with attributes
G.add_nodes_from([
  (1, {"feature": 1, "label": 1}),
  (2, {"feature": 2, "label": 2})
]) #(node, attrdict)

# Loop through all the nodes
# Set data=True will return node attributes
for node in G.nodes(data=True):
  print(node)

# Get number of nodes
num_nodes = G.number_of_nodes()
print("G has {} nodes".format(num_nodes))
```

## Edge

Similar to nodes, edges (with attributes) can also be easily added to NetworkX graphs.

```python
# Add one edge with edge weight 0.5
G.add_edge(0, 1, weight=0.5)

# Get attributes of the edge (0, 1)
edge_0_1_attr = G.edges[(0, 1)]
print("Edge (0, 1) has the attributes {}".format(edge_0_1_attr))
```

```python
# Add multiple edges with edge weights
G.add_edges_from([
  (1, 2, {"weight": 0.3}),
  (2, 0, {"weight": 0.1})
])

# Loop through all the edges
# Here there is no data=True, so only the edge will be returned
for edge in G.edges():
  print(edge)

# Get number of edges
num_edges = G.number_of_edges()
print("G has {} edges".format(num_edges))
```

## Visualization

```python
# Draw the graph
nx.draw(G, with_labels = True)
```

## Node Degree and Neighbor

```python
node_id = 1

# Degree of node 1
print("Node {} has degree {}".format(node_id, G.degree[node_id]))

# Get neighbor of node 1
for neighbor in G.neighbors(node_id):
  print("Node {} has neighbor {}".format(node_id, neighbor))
```

## Random Graph creation and analysis

```python
# creation of a random network with N nodes and E edges - this a erdos-renyi graph
def create(n,m):
    G = nx.gnm_random_graph(n, m, seed = random_state)
    return G
```

```python
def degree_analysis(G):
    degree_sequence = sorted((d for n, d in G.degree()), reverse=True)

    fig = plt.figure("Degree of a random graph", figsize=(8, 8))
    # Create a gridspec for adding subplots of different sizes
    axgrid = fig.add_gridspec(2, 4)


    ax1 = fig.add_subplot(axgrid[:, :2])
    ax1.plot(degree_sequence, "b-", marker="o")
    ax1.set_title("Degree Rank Plot")
    ax1.set_ylabel("Degree")
    ax1.set_xlabel("Rank")

    ax2 = fig.add_subplot(axgrid[:, 2:])
    ax2.bar(*np.unique(degree_sequence, return_counts=True))
    ax2.set_title("Degree histogram")
    ax2.set_xlabel("Degree")
    ax2.set_ylabel("# of Nodes")

    plt.show()
```

```python
random_state = np.random.RandomState(42)
N=20 #number of nodes
E=2*N #number of edges
G=create(N,E)
```

```python
print(G.edges(nbunch=4))
print(G.adj)
print(G.edges)
print(type(G))
print(len(G.edges))
```

```python
pos_cons = nx.spring_layout(G,seed = random_state)
nx.draw(G, pos=pos_cons,labels ={node:node for node in G.nodes()}) #spring_layout , spectral_layout
plt.draw()
```

```python
# Get the adjacency matrix
A = nx.adjacency_matrix(G).toarray()
print("Adjacency Matrix:")
print(A)

# Get the degree matrix

D = np.diag([d for n, d in G.degree()])
print("Degree Matrix:")
print(D)


# Get the Laplacian matrix

L = nx.laplacian_matrix(G).toarray()
print("Laplacian Matrix:")
print(L)
```

```python
degree_analysis(G)
```

```python
# calculate the graph density for the above created random graph

def calculate_density(G):
    n = G.number_of_nodes()
    m = G.number_of_edges()
    # For undirected graph: density = 2m/(n(n-1))
    density = (2 * m) / (n * (n - 1))
    return density

# Calculate density for initial graph (N=20)
initial_density = calculate_density(G)
print(f"Graph density for N=20: {initial_density:.4f}")
```

```python
# Create a larger random graph with N=200
N_large = 200
E_large = 2 * N_large
G_large = create(N_large, E_large)

large_density = calculate_density(G_large)
print(f"Graph density for N=200: {large_density:.6f}")

print(f"\nDensity comparison:")
print(f"N=20: {initial_density:.4f}")
print(f"N=200: {large_density:.6f}")
print(f"Density ratio (N=20/N=200): {initial_density/large_density:.2f}")

def analyze_degree_distribution(G, title):
    degrees = [d for n, d in G.degree()]
    avg_degree = sum(degrees) / len(degrees)
    max_degree = max(degrees)
    min_degree = min(degrees)

    print(f"\n{title}:")
    print(f"Average degree: {avg_degree:.2f}")
    print(f"Max degree: {max_degree}")
    print(f"Min degree: {min_degree}")
    print(f"Number of nodes: {G.number_of_nodes()}")
    print(f"Number of edges: {G.number_of_edges()}")

    return degrees

degrees_20 = analyze_degree_distribution(G, "Degree distribution for N=20")
degrees_200 = analyze_degree_distribution(G_large, "Degree distribution for N=200")

# Density decreases from ~0.105 to ~0.01 as N increases (same E/N ratio)

# Average degree remains ~4 for both (since E = 2N)

# Degree distribution becomes more concentrated around the mean for larger N

# The shape approaches a Poisson distribution – characteristic of Erdős–Rényi random graphs
```

```python
#increase the N from 20 to 200 and observe graph density and degree distribution and explain what you observe

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(degrees_20, bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.title('Degree Distribution (N=20)')
plt.xlabel('Degree')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(degrees_200, bins=30, alpha=0.7, color='red', edgecolor='black')
plt.title('Degree Distribution (N=200)')
plt.xlabel('Degree')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Log-log plots (optional)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
unique, counts = np.unique(degrees_20, return_counts=True)
plt.loglog(unique, counts, 'bo', alpha=0.7)
plt.title('Log-Log Degree Distribution (N=20)')
plt.xlabel('Degree (log)')
plt.ylabel('Frequency (log)')

plt.subplot(1, 2, 2)
unique, counts = np.unique(degrees_200, return_counts=True)
plt.loglog(unique, counts, 'ro', alpha=0.7)
plt.title('Log-Log Degree Distribution (N=200)')
plt.xlabel('Degree (log)')
plt.ylabel('Frequency (log)')

plt.tight_layout()
plt.show()


# Density drops dramatically (from 10.5% to 1%)

# In a party: With 20 people, you can know almost everyone. With 200 people, you only know a small fraction.

# Average connections stay the same (~4 connections per person)

# Whether there are 20 or 200 people at the party, each person still knows about 4 others

# The network becomes more "predictable"

# With 20 people: Some might know 8 people, some only 1

# With 200 people: Almost everyone knows 3-5 people (more consistent)
```
